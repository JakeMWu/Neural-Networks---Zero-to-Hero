We are following Andrej Karpathy's 'spelled-out intro to neural networks and backpropogation: building micrograd'. This first section demonstrates how to accomplish backpropogation by hand. Backpropagation is fundamental in the operation of neural networks.

Essentially our neural network gives us some output as a mathematical expression of all of its connected nodes. We are interested in the gradient of each given node in that network which contributes to this output. We accomplish this iteratively by using the chain rule front to back. If we can tell how the final output or classification has performed and how we might like it to perform we can use our error to alter weight appropriately using backpropogation given our knowledge of these gradients. 

```python
class Value:

    def __init__(self,data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0
        self._prev = set(_children)
        # allows us to see the values immediately composing the parent
        # We also want to track the operations which led to the parent value
        self._op = _op
        self.label = label
    
    # repr formats the expression
    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        out = Value(self.data + other.data, (self, other), '+')
        return out 
    
    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out 
```

```python
a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(10.0, label='c')
e = a*b; e.label = 'e'
d = e + c; d.label = 'd'
f = Value(-2.0, label='f')
L = d * f; L.label = 'L'
L
```

```python
def trace(root):
    # recusively builds a set of all nodes and edges in a graph based 
    # on a given root node
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges
    
def draw_dot(root):
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR = left to right
    
    nodes, edges = trace(root)
    for n in nodes:
        uid = str(id(n))
        # for any value in the graph, create a rectangular ('record') node for it
        dot.node(name = uid, label = "{ %s | data %.4f}" % (n.label,n.data), shape='record')
        if n._op:
            # if this value is a result of some operation create an op node for it 
            dot.node(name = uid+ n._op, label = n._op)
            # and connect this node to it
            dot.edge(uid + n._op, uid)
            
    for n1, n2 in edges:
        #connect n1 to the op node of n2
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)
        
    return dot #alternatively dot.render(view=True)
```

### Backpropagation
We're going to perform backpropogation manually by calculating the gradient, or derivative, of each node with respect to L going backwards using the chain rule. 

Let's consider dL/dd.
L = d * f
therefore 
dL/dd = f


More explicitly from first principles we can consider 
f'(x) = lim h-> 0 (f(x+h)-f(x))/h
The rise over the run as the difference tends to zero
dL/dd = ((L+h) -L)/h 
      = ((d+h)f) - df)/h
      = (df + hf - df)/h
      = hf/h
      = f
